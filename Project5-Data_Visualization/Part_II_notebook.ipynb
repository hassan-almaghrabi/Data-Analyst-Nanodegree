## Investigation Overview


> In this investigation, I wanted to take a look at how the duration of trips changes with respect to day of week and customer type and gender. 


## Dataset Overview

> I have used Ford GoBike Dataset, which is a dataset data set includes information about individual rides made in a bike-sharing system covering the greater San Francisco Bay area, contaian about 183412 observation with 16 different features.

````python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
````

````python
# load in the dataset into a pandas dataframe
df = pd.read_csv('201902-fordgobike-tripdata.csv')

# converting the start and end time to datetime type
df[['start_time', 'end_time']] = df[['start_time', 'end_time']].astype('datetime64[ns]')

# Extract day and hour from the start time of the ride
df['start_day'] = df['start_time'].dt.day_name()
df['start_hour'] = df['start_time'].dt.hour
````

## Trips per Weekdays


> Most trips were taken on Thursdays, and the weekend have the lowest trips count.

````python
plt.figure(figsize=(8,6))
base_color = sb.color_palette()[0]
days = ['Monday','Tuesday','Wednesday','Thursday','Friday', 'Saturday', 'Sunday']
sb.countplot(data=df, x='start_day', order=days, color=base_color)
plt.title('Trips in days of week')
plt.xlabel('day of week');
````

## Distribution of Rides Duration by User Types

> The plot shows the distribution for different bike user type ,It appears that the customer user type takes a longer trips duration than the subscriber user type.

````python
# Set the limit of trips duration to 3000 second to remove the extreme values
filter_data = df.query('duration_sec < 3000')

plt.figure(figsize=(6,7))
sb.violinplot(data=filter_data, x='user_type', y='duration_sec', color=base_color, inner='quartile')
plt.title('Duration in Second per User Type')
plt.xlabel('User Type')
plt.ylabel('Duration (sec)');
````

## Gender Average Duration in Second per User Type

> Females users take longer trips durations than males no matter what the user type.

````python
plt.figure(figsize=(8,6))
sb.pointplot(data=df, x='member_gender', y='duration_sec', linestyles='', hue='user_type', dodge=0.2)
plt.title('Gender Avg. Duration per User Type')
plt.ylabel('Avg. Duration (sec)');
````
